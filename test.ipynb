```python
import random
list1 = random.sample(range(3,51), 2)
def huzi(a,b):
    print("%d and %d 是否互質?" % (a,b))
    for i in range(2,min(a,b)+1):
        if a%i==0 and b%i==0:
            return ("NO")
    return ("YES")
print(huzi(list1.pop(),list1.pop()))
```

```python
#-*- coding=utf-8 -*-
import io
import csv
import pandas
dfs = pandas.read_html('http://www.taiwanoil.org/')
#pandas的read_html是一種dataFrame(類似表格的格式)，可以自行選擇要的row跟columns
type(dfs)
len(dfs)
gasPrice = dfs[1]
gasPrice = gasPrice.iloc[7:15,0:5]
gasPrice.columns = [u'油別',u'今日油價',u'施行日期',u'下周批發價(%)',u'預測下周漲跌(元)']
#gasPrice['油別'] = 1 此方法只能插入在最後一列
gasCompanyArray = [u'中油',u'中油',u'中油',u'中油',u'台塑',u'台塑',u'台塑',u'台塑']
gasStateArray = [u'98',u'95',u'92',u'柴油',u'98',u'95',u'92',u'柴油']
gasPrice.insert(0,u'供應商',gasCompanyArray)  #insert(插入第幾列,列名稱,value)
gasPrice[u'油別'] = gasStateArray
#前方加個u是將內容轉為unicode格式，避免後續存檔轉檔時發生亂碼
gasPrice.to_csv('gasPrice.csv',index=False,encoding="big5")
gasPrice
```
