```python
# -*- coding: UTF-8 -*-
#ā ē ī ō ū 1  
# á é í ó ú 2 
#ǎ ě ǐ ǒ ú(ǚ) 3
# à è ì ò ù(ǜ) 4
#https://www.ifreesite.com/pinyin/
#中文檔名=單個數字編號, 拼音檔名=單個數字編號-1, 輸出檔案=數字編號_簡立軒
#       ex:9                 ex:9-1                 ex:9_簡立軒
#修改id_Num參數即可
#須注意'不','分'等破音字的發音,需自行修正一二三四聲
#遇到數字或特殊符號會產生空格,待debug或輸出後自行修正
#讀取寫入 with open

path_='C:/Users/cow81/多音字－單字語料/多音字－單字語料'
id_Num = input('請提前翻譯好拼音並存入"編號-1.txt",再輸入編號:')
with open(path_+'/{}-1.txt'.format(id_Num), 'r+',encoding='utf-8', newline='') as txtfile,open(path_+'/{}.txt'.format(id_Num), 'r+',encoding='utf-8', newline='') as txtfile2,open(path_+'/{}_簡立軒.txt'.format(id_Num), 'w+',encoding='utf-8', newline='') as txtfile3:
    while(True):
        line1 = txtfile.readline().strip('\n')
        line1CHS = txtfile2.readline()
        #print(line1CHS)
        line1CHS = line1CHS.strip('\r\n')+':'
        #print(line1CHS)
        list1 = line1.split(' ')
        listAdd = []
        listAdd.append(line1CHS)
        lenChs = len(line1CHS)-1
        for chs in line1CHS:
            if(chs=='？' or chs=='（' or chs=='）' or chs=='“' or chs=='0' or chs=='1' or chs=='2' or chs=='3' or chs=='4' or chs=='5' or chs=='6' or chs=='7' or chs=='8' or chs=='9'):
                lenChs = lenChs - 1
        #print(lenChs)
        #print(listAdd)
        if(len(list1)<2):
            break
        #print(len(list1))
        for s in list1:
            s1 = ''
            numB = ''
            #print(s)
            for c in s:
                #if(c=='？' or c=='（' or chs=='）' or c=='“' or c=='0' or c=='1' or c=='2' or c=='3' or c=='4' or c=='5' or c=='6' or c=='7' or c=='8' or c=='9'):
                    #continue
                if(c=='ā' or c=='á' or c=='ǎ' or c=='à'):
                    s1=s1+'a'
                if(c=='ē' or c=='é' or c=='ě' or c=='è'):
                    s1=s1+'e'
                if(c=='ī' or c=='í' or c=='ǐ' or c=='ì'):
                    s1=s1+'i'
                if(c=='ō' or c=='ó' or c=='ǒ' or c=='ò'):
                    s1=s1+'o'
                if(c=='ū' or c=='ú' or c=='ǔ' or c=='ù' or c=='ǚ' or c=='ǜ'):
                    s1=s1+'u'
                if(c=='“' or c=='（' or c=='）' or c=='？'):
                    c=''
                if(c=='ā' or c=='ē' or c=='ī' or c=='ō' or c=='ū'):
                    numB = '1'
                    continue
                if(c=='á' or c=='é' or c=='í' or c=='ó' or c=='ú'):
                    numB = '2'
                    continue
                if(c=='ǎ' or c=='ě' or c=='ǐ' or c=='ǒ' or c=='ǔ'):
                    numB = '3'
                    continue
                if(c=='à' or c=='è' or c=='ì' or c=='ò' or c=='ù'):
                    numB = '4'
                    continue
                s1 = s1 + c
            s1 = s1+numB
            if(len(listAdd)<=lenChs):
                listAdd.append(s1)
        #print(listAdd)
        pnt = ''
        for p in listAdd:
            pnt = pnt+p+' '
        print(pnt)
        txtfile3.write(pnt+'\n')
txtfile.close()
txtfile2.close()
txtfile3.close()  #write本身自帶close,此行可有可無,影響不大
```
