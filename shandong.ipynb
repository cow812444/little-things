```python
# -*- coding: UTF-8 -*-
import cv2 as cv
import cv2
import numpy as np
import os
 
def do_image(src001,src002,saveName):
    x, y = src001.shape[0:2]
    n1 = cv.resize(src001, (int(y / 4 * 1.08), int(x / 4 * 1.08)))
    x, y = src002.shape[0:2]
    n2 = cv.resize(src002, (int(y / 4 * 1.08), int(x / 4 * 1.08)))
    mix02 = np.concatenate([n2, n1] , axis=1)  #橫向合併
    cv2.putText(mix02, '69', (600, 50), cv2.FONT_HERSHEY_SIMPLEX,  2, (0, 0, 255), 3, cv2.LINE_AA)
    cv2.putText(mix02, '68', (720, 50), cv2.FONT_HERSHEY_SIMPLEX,  2, (0, 0, 255), 3, cv2.LINE_AA)
    #cv.imshow("mix", mix02)
    cv.imwrite(path_save+saveName,mix02)


path_001='C:/Users/cow81/LabelingTool/shandong/pic2/pic/7-3/68/'
path_002='C:/Users/cow81/LabelingTool/shandong/pic2/pic/7-3/69/'
path_save='C:/Users/cow81//LabelingTool/shandong/pic2/pic/7-3/mixed/'
files_001=os.listdir(path_001)
files_002=os.listdir(path_002)
#print(files_001) 

n=0 #設定初始值
for i in files_001:
    src001 = cv.imread(path_001+i)
    src002 = cv.imread(path_002+files_002[n])
    fileName = i.split('_')
    fileName1 = []
    for p in range(1,8):
        fileName1.append(fileName[p])
    #print(fileName1)
    saveName = ''
    for s in fileName1:
        saveName = saveName + s
        if(len(s)<=4):
            saveName = saveName + '_'
    print(saveName)
    do_image(src001,src002,saveName)
    n=n+1
cv.waitKey(0)
cv.destroyAllWindows()
```
